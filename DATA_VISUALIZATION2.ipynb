```python
country_list=list(world_hapiness_2016["Country"].unique())
happiness_score_list=list(world_hapiness_2016["Happiness Score"])
data=pd.DataFrame({"country_list":country_list,"happiness_score_list":happiness_score_list})
new_index=(data["happiness_score_list"].sort_values(ascending=False)).index.values
sorted_data=data.reindex(new_index)
sorted_data
#ascending=False means my data would be in descending order.azalan sırada olsun
#ascendin=True means will be in ascending order.artan sırada olur.
```

```python
#lets virtualize data 
plt.figure(figsize=(50,30))
sns.barplot(x=sorted_data["country_list"],y=sorted_data["happiness_score_list"])
plt.xticks(rotation=90)
plt.xlabel("Countries")
plt.ylabel("country hapiness score")
plt.title("World Happiness Report")
plt.show()
```
